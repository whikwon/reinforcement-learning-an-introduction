### Mountain Car

1. 문제: 자동차가 bowl 형태의 길에서 움직이고 있다. 가속, 감속을 통해서 오른쪽 끝에 도달하는 것이 목표이다.
2. 조건
    - state: 자동차의 위치($x_t$)와 속도($\dot x_t$)로 이루어진 *tuple* 
        - 왼쪽 끝(-1.2)에 도달 시 $x_t$가 0으로 reset 됨.
        - $x_t \doteq \textit{bound}[x_t + \dot x_{t+1}]$와 $\dot x_{t+1} \doteq \textit{bound}[\dot x_t + 0.001 A_t - 0.0025cos(3x_t)$ 식에 의해 위치와 속도가 결정 됨.
        - 속도 제한은 -0.07~0.07이며 벗어나는 경우 bound 값을 갖게 됨.        
    - action: 앞으로 가속(+1), 뒤로 가속(-1), 가속 없음(0)
    - reward: 모든 time step(-1), 꼭대기에 도달 시(0)
    - terminal: 오른쪽 끝(0.5)
3. 학습 목표
    - Tile Coding
    - Episodic Semi-gradient Sarsa

```python
import matplotlib.pyplot as plt
import numpy as np
from TileCoding import *
from math import cos
from random import uniform
from itertools import count
```

```python
ACTION_REVERSE = -1
ACTION_ZERO = 0
ACTION_FORWARD = 1
ACTIONS = [ACTION_REVERSE, ACTION_ZERO, ACTION_FORWARD]

# bound for position and velocity
POSITION_MIN = -1.2
POSITION_MAX = 0.5
VELOCITY_MIN = -0.07
VELOCITY_MAX = 0.07
```

```python
class MountainCar:
    
    def __init__(self):
        self.reset()
        self.nA = 3
        
    def step(self, action):
        done = False
        reward = -1
        self.velocity = self.velocity + 0.001 * action - 0.0025 * np.cos(3 * self.position)
        self.velocity = max(min(self.velocity, VELOCITY_MAX), VELOCITY_MIN)
        self.position += self.velocity
        self.position = max(min(self.position, POSITION_MAX), POSITION_MIN)
        if self.position == POSITION_MIN:
            self.velocity = 0 
        elif self.position == POSITION_MAX:
            done = True
            reward = 0
        return (self.position, self.velocity), reward, done
    
    def reset(self):
        self.position = uniform(-0.6, -0.4)
        self.velocity = 0 
        return (self.position, self.velocity)
        
```

Tile Coding에 대한 내용은 일단 넘어가도록 하자. 대충 로직은 tile을 2차원 공간에서 16개씩 각각 쪼갠다. 그리고, 현재 state(position, velocity)에 대해서 8개의 tile 위치를 확인하고 값을 반환해준다. 현재 state에 대해서 tiling 별로 어떤 tile들이 영향을 미치는 지 알 수 있기 때문에 해당 가중치에 대해서 학습한다.

```python
class ValueFunction:
    # In this example I use the tiling software instead of implementing standard tiling by myself
    # One important thing is that tiling is only a map from (state, action) to a series of indices
    # It doesn't matter whether the indices have meaning, only if this map satisfy some property
    # View the following webpage for more information
    # http://incompleteideas.net/sutton/tiles/tiles3.html
    # @maxSize: the maximum # of indices
    # max_size가 2048이라는 건 전체 tile에 대한 weight 수가 2048개라는 것이고 tiling이 8개니까 space를 dim당 16개로 쪼갠거다. 
    def __init__(self, step_size, num_tilings=8, max_size=2048):
        self.max_size = max_size
        self.num_tilings = num_tilings

        # divide step size equally to each tiling
        self.step_size = step_size / num_tilings

        self.hash_table = IHT(max_size)

        # weight for each tile
        self.weights = np.zeros(max_size)

        # position and velocity needs scaling to satisfy the tile software
        # scale 후에 min ~ max 값의 범위가 num_tilings 로 조정된다. 
        self.position_scale = self.num_tilings / (POSITION_MAX - POSITION_MIN)
        self.velocity_scale = self.num_tilings / (VELOCITY_MAX - VELOCITY_MIN)

    # get indices of active tiles for given state and action
    def get_active_tiles(self, position, velocity, action):
        # I think positionScale * (position - position_min) would be a good normalization.
        # However positionScale * position_min is a constant, so it's ok to ignore it.
        active_tiles = tiles(self.hash_table, self.num_tilings,
                            [self.position_scale * position, self.velocity_scale * velocity],
                            [action])
        return active_tiles

    # estimate the value of given state and action
    def value(self, position, velocity, action):
        if position == POSITION_MAX:
            return 0.0
        active_tiles = self.get_active_tiles(position, velocity, action)
        return np.sum(self.weights[active_tiles])

    # learn with given state, action and target
    def learn(self, position, velocity, action, target):
        active_tiles = self.get_active_tiles(position, velocity, action)
        estimation = np.sum(self.weights[active_tiles])
        delta = self.step_size * (target - estimation)
        for active_tile in active_tiles:
            self.weights[active_tile] += delta

    # get # of steps to reach the goal under current state value function
    def cost_to_go(self, position, velocity):
        costs = []
        for action in ACTIONS:
            costs.append(self.value(position, velocity, action))
        return -np.max(costs)
```

```python
# policy prob 구하고 pick 하는 것까지. 
def epsilon_greedy_policy(Q_obs, epsilon, nA):
    policy = np.ones(nA, dtype=float) * epsilon / nA
    best_action = np.argmax(Q_obs)
    policy[best_action] += 1.0 - epsilon
    return np.random.choice(ACTIONS, p=policy)
```

### Semi-Gradient Sarsa
![episodic_semi_gradient_sarsa](../images/episodic_semi_gradient_sarsa.png)
<center> 출처: Reinforcement Learning : An Introduction. Richard S. Sutton and Andrew G. Barto. 2017 (pp. 199) </center>

```python
def semi_gradient_sarsa(env, num_episodes, value_fn, policy_fn, discount_factor=0.95, epsilon=0.1):
    counter = []
    for i_episode in range(num_episodes):
        print("Episode {} On-going".format(i_episode+1), end='\r')
        position, velocity = env.reset()
        for t in count():    
            value = [value_fn.value(position, velocity, a) for a in ACTIONS]        
            action = policy_fn(value, epsilon, env.nA)
            (next_position, next_velocity), reward, done = env.step(action)        
            if done:
                value_fn.learn(position, velocity, action, reward)
                counter.append(t + 1)
                break
            next_value = [value_fn.value(next_position, next_velocity, a) for a in ACTIONS]
            next_action = policy_fn(value, epsilon, env.nA)
            target = reward + discount_factor * value_fn.value(next_position, next_velocity, next_action)
            value_fn.learn(position, velocity, action, target)
            position, velocity = next_position, next_velocity             
    return counter
```

```python
env = MountainCar()
policy_fn = epsilon_greedy_policy
step_size = [0.1, 0.2, 0.5]
for s in step_size:
    value_fn = ValueFunction(s)
    counter = semi_gradient_sarsa(env, 100, value_fn, policy_fn)
    plt.plot(counter, label='alpha={}/8'.format(s))
    plt.yscale('log')
    plt.ylim([100, 1000])
    plt.xlabel('Episode')
    plt.ylabel('Steps per episode')
    plt.legend()
plt.show()    
```

```python
position, velocity = env.reset()
p, v = [position], [velocity]
while True:
    value = [value_fn.value(position, velocity, a) for a in ACTIONS]
    action = epsilon_greedy_policy(value, 0.1, env.nA)
    (next_position, next_velocity), reward, done = env.step(action)
    if done:
        break
    position, velocity = next_position, next_velocity
    p.append(position), v.append(velocity)
```

```python
plt.plot(p)
plt.title('Car position changes after learning')
plt.xlabel('time steps')
plt.ylabel('position')
plt.show()
```

### Semi-gradient n_step Sarsa
![episodic_semi_gradient_nstep_sarsa](../images/episodic_semi_gradient_nstep_sarsa.png.png)
<center> 출처: Reinforcement Learning : An Introduction. Richard S. Sutton and Andrew G. Barto. 2017 (pp. 200) </center>

```python
def semi_gradient_n_step_sarsa(env, num_episodes, value_fn, policy_fn, discount_factor=0.95, epsilon=0.1, n=2):
    counter = []
    for i_episode in range(num_episodes):
        print("Episode {} On-going".format(i_episode+1), end='\r')
        position, velocity = env.reset()
        value = [value_fn.value(position, velocity, a) for a in ACTIONS]
        action = policy_fn(value, epsilon, env.nA)
        actions_sum = [action]
        positions_sum = [position]
        velocities_sum = [velocity]
        rewards_sum = [0.0]                    
        T = float('inf')
        t = 0
        while True:  
            if t < T:
                (next_position, next_velocity), reward, done = env.step(action)
                positions_sum.append(next_position)
                velocities_sum.append(next_velocity)
                rewards_sum.append(reward)
                if done:
                    T = t + 1
                else:
                    value = [value_fn.value(next_position, next_velocity, a) for a in ACTIONS]
                    next_action = policy_fn(value, epsilon, env.nA)
                    actions_sum.append(next_action)
            tau = t - n + 1 
            if tau >= 0:
                G = sum(pow(discount_factor, i-tau-1) * rewards_sum[i] for i in range(tau+1, min(tau+n+1, T+1)))
                if tau + n < T:
                    G += pow(discount_factor, n) * value_fn.value(
                         positions_sum[tau+n], velocities_sum[tau+n], actions_sum[tau+n])
                value_fn.learn(positions_sum[tau], velocities_sum[tau], actions_sum[tau], G)                             
            if tau == T - 1:
                break                            
            position, velocity, action = next_position, next_velocity, next_action
            t += 1
        counter.append(t)
    return counter
```

### Figure 10.3
아래 조건에 대한 학습 비교 결과이다.
- $n=1,\ \alpha=0.5/8$
- $n=8,\ \alpha=0.3/8$ 

```python
step_size = 0.1
env = MountainCar()
policy_fn = epsilon_greedy_policy

for n, a in zip([1, 8], [0.5, 0.3]):
    value_fn = ValueFunction(step_size)
    counter = semi_gradient_n_step_sarsa(env, 500, value_fn, policy_fn, 0.95, 0.1, n)
    plt.plot(counter, label='n={}'.format(n))
    plt.legend()
    plt.xlabel('Episode')
    plt.ylabel('Steps per episode')
    plt.yscale('log')
plt.show()
```
